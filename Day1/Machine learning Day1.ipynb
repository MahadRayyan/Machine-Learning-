```python
df=pd.read_csv(r"C:\Users\Ruhan\Desktop\QWETRUM task1- Syed MAhad Ur Rayyan\Messy_Employee_dataset.csv")
df.head(3)
```

```python
# if we want to replace null values with random value
df.fillna(10)
```

```python
#backward data is forward to next one
df.fillna(method='bfill',axis=0)
#forward data is backward to previous one
df.fillna(method='ffill',axis=1)
#if axis is 0 row wise and if axis =1 then column wise
```

```python
#replace object type null values with mode value (most repeated) for single column
#df.info()#give info about dataset
df["Performance_Score"].fillna(df["Performance_Score"].mode()[0],inplace=True)
```

```python
#replace object type null values with mode value (most repeated) for multiple column
```

```python
for i in df.select_dtypes(include="object").isnull().columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
# replacing numeric values 
# can be replaced through mean median mode and any constant value 
for i in df.select_dtypes('float64').isnull().columns:
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
df.select_dtypes("float64").columns
```

```python
#removing missing values through sklearn
from sklearn.impute import SimpleImputer
si =SimpleImputer(strategy='mean')
ar=si.fit_transform(df[['Age', 'Salary']])
```

```python
newdf= pd.DataFrame(ar,columns=df.select_dtypes("float64").columns)
newdf.head()
```
