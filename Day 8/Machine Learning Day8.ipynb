logistic regression: when we have to predict discrete output like (yes or no) we use this algo its uses sigmoid function which have threshold 0.5 and its formula is y^=(1/1+e^-x) where x formula is same as simple linear regression m1x1+m2x2+b


# binomial logistic regression
when there is only one independent variable 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("netflix_age_dataset.csv")
dataset.head(3)
```

```python
dataset.shape
```

```python
plt.figure(figsize=(4,2))
sns.scatterplot(x=dataset["age"],y=dataset["netflix"],hue=dataset["netflix"])
plt.show()
```

```python
x=dataset[["age"]]
y=dataset["netflix"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict(x_test)
```

```python
plt.figure(figsize=(4,2))
sns.lineplot(x=dataset["age"],y=lr.predict(x))
sns.scatterplot(x=dataset["age"],y=dataset["netflix"],hue=dataset["netflix"])
plt.show()
```

# Multinomial logistic regression 
when there is more than one independent variable

```python
dataset=pd.read_csv("netflix_dataset_2_features.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x=dataset["age"],y=dataset["monthly_income"],hue=dataset["netflix"])
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["netflix"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
