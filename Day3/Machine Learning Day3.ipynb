# Feature Scaling

# standardization

by feature scaling nature of the data remain same but magnitude is reduce it is done so that all the entity of dataset have same magnitude by standardization mean of the data become 0 and var 1 formula for this is Xnew=(Xi-Xmean)/std

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\Ruhan\Desktop\QWETRUM task1- Syed MAhad Ur Rayyan\Messy_Employee_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["Salary"].fillna(df["Salary"].mean(),inplace=True)
```

```python
ss=StandardScaler()
ss.fit(df[["Salary"]])
```

```python
df["Salaryss"]=pd.DataFrame(ss.transform(df[["Salary"]]),columns=["x"])
```

```python
df.head()
```

```python
df.describe()
```

# Normalization

 by Normalization data is scaled between 0 and 1 formula for this is Xnew=(Xi-Xmean)/(MAXi-MINi)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm=MinMaxScaler()
mm.fit(df[["Salary"]])
```

```python
df["Salarymm"]=pd.DataFrame(mm.transform(df[["Salary"]]),columns=["x"])
```
